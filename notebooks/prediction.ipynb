### Tratando os dados

```python
deaths.rename(columns={'Province/State': 'State', 'Country/Region': 'Country'}, inplace=True)
```

```python
deaths.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
deaths.head(2)
```

```python
deaths = deaths.groupby('Country').sum().reset_index()
```

```python
deaths.head(2)
```

Vamos pivotar a tabela para que as datas fiquem em apenas uma coluna.

```python
# Definindo as colunas que quero pivotar
cols_p = list(deaths.columns)
cols_p.remove('Country')

# Pivotando as colunas
deaths_2 = pd.DataFrame(deaths.set_index('Country').stack()).reset_index()
deaths_2.columns = ['country', 'date', 'qtd']

# Mais alguns tratamentos
deaths_2['date'] = pd.to_datetime(deaths_2['date'])
deaths_2.sort_values(by=['country', 'date'], inplace=True)

deaths_2.head()
```

Vamos filtrar os dados a partir das datas em que ocorreu o primeiro caso em cada pas.

```python
# Vou armazenar em uma lista todos os ndices que quero remover
countries = deaths_2['country'].unique()
# countries = ['Afghanistan', 'Algeria']
list_remove = list()

for c in countries:
    df_filtered = deaths_2[deaths_2['country'] == c]
    lista_qtd = df_filtered['qtd'].values.tolist()

    first_index_death = next((i for i, x in enumerate(lista_qtd) if x), 0)

    if first_index_death > 0:
        i0 = df_filtered.index[0]
        i1 = i0 + first_index_death
        list_extend = list(range(i0, i1))
        list_remove.extend(list_extend)
```

```python
deaths_2.iloc[list_remove]['qtd'].sum() # Dar 0 aqui no garante que est correto, mas se der diferente de 0, com certeza est errado.
```

```python
# Removo os ndices da lista que criei logo acima
deaths_3 = deaths_2.iloc[~deaths_2.index.isin(list_remove)].reset_index(drop=True)
```

```python
deaths_3.head()
```

### Modelagem

Inicialmente, vamos criar um modelo para um nico pas. Aleatoriamente, sem qualquer tipo de vis, vamos escolher o Brasil.

```python
brazil = deaths_3[deaths_3['country'] == 'Brazil'].reset_index(drop=True)
```

```python
brazil.head(2)
```

```python
brazil.drop('country', axis=1, inplace=True)
brazil.columns = ['ds', 'y']
brazil.head(2)
```

```python
brazil.shape
```

```python
def mape(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return mape
```

```python
def forecast(data, metric, test_size=60, prediction_size=7):
    # Resetando os ndices dos dados
    data.reset_index(drop=True, inplace=True)

    # Definindo o ndice de treinamento
    final_index_train = len(data) - (prediction_size + 1)
    index_train = final_index_train - prediction_size

    # Criando listas para armazenar os resultados
    y_real = list()
    y_pred = list()

    for i in range(test_size):
        # Definindo os dados de treino
        data_train = data[:-index_train]
        
        # Treinando o modelo e escorando os dados de validao
        m = Prophet()
        m.fit(df=data_train)
        future = m.make_future_dataframe(periods=prediction_size)
        forecast = m.predict(future.tail(prediction_size))

        # Appendando os valores nas listas
        y_real.append(data[-index_train:].head(n=prediction_size)['y'].sum())
        y_pred.append(forecast.tail(n=7)['yhat'].sum())

        index_train-=1
    
    # Calculando a performance
    result = metric(y_real, y_pred)  # Funo definida anteriormente

    return result
```

```python
resultado = forecast(data=brazil, metric=mape, test_size=60, prediction_size=7)
```

```python
print('MAPE: {:.2f}'.format(resultado))
```

Agora, precisamos criar uma funo que vai receber os dados com todos os pases e, para cada um deles, ajustar um modelo.

```python
def forecast_all(data, col_sep):
    
    df_results = pd.DataFrame(columns=['entidade', 'resultado'])
    items = data[col_sep].unique().tolist()
    # items = ['Brazil', 'Afghanistan']

    for i in items:
        data_filtered = data[data[col_sep] == i].copy()
        data_filtered.drop(col_sep, axis=1, inplace=True)
        data_filtered.columns = ['ds', 'y']
        result = forecast(data=data_filtered, metric=mape, test_size=30, prediction_size=7) # Funo definida anteriormente

        df_results = df_results.append({'entidade': i, 'resultado': result}, ignore_index=True)
    
    return df_results
```

```python
results = forecast_all(data=deaths_3, col_sep='country')
```
